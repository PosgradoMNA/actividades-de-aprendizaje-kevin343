### Descomposición

Encuentra los eigenvalores y eigenvectores de las siguientes matrices y reconstruye la matriz original a traves de las matrices  WDW−1  (OJO. Esto es lo mismo de la ecuación del paso 2 solo le cambiamos la variable a la matriz diagonal)

```python
import numpy as np
from numpy import array
from numpy.linalg import eig
```

```python
#Matriz A
A = array([[3, 0, 2], [3, 0, -2], [0, 1, 1]])
print("-------Matriz original-------")
print(A)
print("-----------------------------")
```

```python
#Obtener eigenvalues y eigenvectors de matriz A
v,ve=eig(A)
print( v )# eigenvalues D
print( ve )#eigenvectors W
```

```python
#Reconstruir matrices originales 

Winv = np.linalg.inv(ve)  #Inversa de W
D = np.diag(v)            #Diagonal
```

```python
X = np.dot(ve,D)   #Operacion de X = W • D
Y = np.dot(X,Winv) #Operacion de  Y = X • W^-1
Rematrix = Y.round()

```

### Matriz A reconstruida


```python
Rematrix        #Matriz A reconstruida
```

```python
B = array([[1, 3, 8], [2, 0, 0], [0, 0, 1]])
print("-------Matriz original-------")
print(B)
print("-----------------------------")
```

```python
v2,ve2=eig(B)
print( v2 )# eigenvalues D
print( ve2 )#eigenvectors W
```

```python
#Reconstruir matrices originales 

Winv2 = np.linalg.inv(ve2)  #Inversa de W
D2 = np.diag(v2)            #Diagonal
```

```python
X2 = np.dot(ve2,D2)   #Operacion de X2 = W • D2
Y2 = np.dot(X2,Winv2) #Operacion de  Y2 = X2 • W^-1
Rematrix2 = Y2.round()
```

### MATRIZ RECONSTRUIDA A2


```python
Rematrix2        #Matriz A2 reconstruida
```

```python
C = array([[5, 4, 0], [1, 0, 1], [10, 7, 1]])
print("-------Matriz original-------")
print(C)
print("-----------------------------")
```

```python
v3,ve3=eig(C)
print( v3 )# eigenvalues
print( ve3 )#eigenvectors
```

```python
#Reconstruir matrices originales 

Winv3 = np.linalg.inv(ve3)  #Inversa de W
D3 = np.diag(v3)            #Diagonal
```

```python
X3 = np.dot(ve3,D3)   #Operacion de X3 = W • D2
Y3 = np.dot(X3,Winv3) #Operacion de  Y3 = X3 • W^-1
Rematrix3 = Y3.round()
```

### MATRIZ RECONSTRUIDA A3


```python
Rematrix3        #Matriz A3 reconstruida
```

## Ejercicio 2

Juega con Lucy, una cisne, ayudala a encontrar cuantos valores singulares necesita para no perder calidad a través de SVD. Posteriormente usa 3 imágenes de tu preferencia y realiza la misma acción :D

A esto se le llama compresión de imagenes :o

```python
from six.moves import urllib
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('classic')
img = Image.open(urllib.request.urlopen('https://biblioteca.acropolis.org/wp-content/uploads/2015/03/Cisne.jpg')).convert('LA')
#img = Image.open('lucy.jpg')
imggray = img.convert('LA')
imgmat = np.array(list(imggray.getdata(band=0)),float)

print(imgmat)

imgmat.shape = (imggray.size[1],imggray.size[0])

plt.figure(figsize=(9,6))
plt.imshow(imgmat,cmap='gray')
plt.show()
print(img)
```

```python
U,D,V = np.linalg.svd(imgmat)
imgmat.shape
```

```python
U.shape
```

```python
V.shape
```

```python
#Cuantos valores crees que son necesarios?
#A=U*D*V
#aqui los elegiremos---------------------------
# por las dimensiones de este caso en particular 
#iremos de 0-660, siendo 660 como normalmente están los datos
#con 50 podemos observar que Lucy se ve casi igual, es decir conservamos aquello que en 
# realidad estaba aportando a la imagen en este caso :D por medio de la variabilidad 
#juega con el valor nvalue y ve que pasa con otros valores
nvalue = 50
#------------------------------
reconstimg = np.matrix(U[:,:nvalue])*np.diag(D[:nvalue])*np.matrix(V[:nvalue,:])
#ve las dimensiones de la imagen y su descomposicion
#660x1024= U(660X660)D(660X1024)V(1024x1024)
        #=U(660Xnvalues)D(nvaluesXnvalue)V(nvaluesx1024)

        #=U(660X50)(50X50)(50X1024)
plt.imshow(reconstimg,cmap='gray')
plt.show()
print("Felicidades la imagen está comprimida")
```

¿Qué es feature importance y para que nos sirve?

RESP : Feature importance se refleja en la magnitud de los valores correspondientes en los vectores propios (mayor magnitud, mayor importancia).Cuanto mayores son estos valores absolutos, más contribuye una característica específica a ese componente principal.
