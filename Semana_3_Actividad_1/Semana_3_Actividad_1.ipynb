## Parte 1: Fundamentos de bases de datos

### Fundamentos de bases de datos


#### Fundamentos de bases de datos y para ciencia de datos.  

Una base de datos consiste en almacenar de forma organizada datos respectos a varios temas de interés en la nube o computador. Este almacenamiento de datos es un lugar donde la información llega y son depositados. Los datos son de gran interés por que como se vio en las actividades pasadas con ellos se puede tomar decisiones con ayuda de otras herramientas de inteligencia artificial para mantener competitividad, monitorear desempeños en las empresas.

Un almacenamiento de datos esta compuesto por capas o niveles : 
    • Los que presentan datos mediante informes
    • El motor de análisis para obtener acceso a los datos y analizarlos 
    • El servidor de datos donde se almacena los datos
Los datos se organizan en columnas y filas.



#### Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

Un data Warehouse es un almacén de datos que recibe información de varias fuentes. La información obtenida es muy valiosa por que sirve para tomar decisiones, son estables y no cambian.
El data warehouse puede almacenarse en la nube para abaratar costes y facilitar su uso. Estos incluso pueden ser autónomos y administrarse solos.
Para crear un datawarehouse se necesita: 
    
   • Contenido de datos específicos
   • Relaciones dentro y entre grupos de datos
   • El entorno de sistemas que soportará el almacén de datos.
   • Los tipos de transformaciones de datos necesarios
   • Frecuencia de actualización de datos

```python
import matplotlib.pyplot as plt   #Revisar la edad por box plot para ver si se puede aplicar la media,mediana en vez de 'Nan'
plt.figure(figsize=(15,3))
dataset.boxplot(column=[ 'Age'],vert=False)
plt.xlim([0, 100])
```

```python

dataset['Age'].median()

```

```python

dataset['Age'].fillna(value=dataset['Age'].mean(), inplace=True)#Reemplazar los Nan por la mediana de la columna Age
```

```python
dataset = dataset.dropna(axis=0, subset=['Y','Gender','Education','Marital status'])#Eliminar las filas que tiene "Nan"

dataset.isna().any()#Verificar que no falta datos con "Nan" en 'Y','Gender','Education','Marital status' y 'Age'
```

```python
dataset = dataset[(dataset["Age"] > 30) & (dataset["Age"] < 50)] # Selecciono las personas entre 30 y 50 años
dataset
```
