#### Q1: Based on the example above, replace NaN in "stroke" column with the mean value.

```python
dff.replace("?", np.nan, inplace = True)
```

```python
# Calcular la media
avg_stroke = dff["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)

# Reemplazar Nan 
dff["stroke"].replace(np.nan, avg_stroke, inplace = True)
```
