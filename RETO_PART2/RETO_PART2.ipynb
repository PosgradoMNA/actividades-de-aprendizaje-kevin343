```python
# Convertir datos categóricos en variables ficticias o indicadoras
y = pd.get_dummies(df.SEMAFORO)
y
```

```python
# Seleccinar variables independientes X y dependiente Y

features = df.drop(['SEMAFORO'],axis=1)


X = pd.get_dummies(features)
```

```python
# Separar la data en 80% para entrenar y 20% para hacer pruebas
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = 0.2, 
                                                    random_state = 0)
```

```python
#Realizar un analisis de las feature importances 
```

```python
# Función para hacer plot de las features mas importantes

def feature_plot(importances, X_train, top_k=5):
    
    # Display the five most important features
    indices = np.argsort(importances)[::-1]
    columns = X_train.columns.values[indices[:top_k]]
    values = importances[indices][:top_k]

    # Creat the plot
    fig = plt.figure(figsize = (9,5))
    plt.title("Normalized Weights for First Five Most Predictive Features", fontsize = 16)
    plt.bar(np.arange(top_k), values, width = 0.6, align="center", color = '#00A000', \
          label = "Feature Weight")
    plt.bar(np.arange(top_k) - 0.3, np.cumsum(values), width = 0.2, align = "center", color = '#00A0A0', \
          label = "Cumulative Feature Weight")
    plt.xticks(np.arange(top_k), columns, rotation= 45)
    plt.xlim((-0.5, top_k-0.5))
    plt.ylabel("Weight", fontsize = 12)
    plt.xlabel("Feature", fontsize = 12)
    
    plt.legend(loc = 'upper center')
    plt.tight_layout()
    plt.show()  
```

```python
#  Feature importance atraves de Random Forest
TOP_K =20
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
# Definir el modelo
model = RandomForestClassifier(random_state = 0,n_estimators=50, max_depth= 10 )
# fit del modelo
model.fit(X_train, y_train)
# get importance
importance = model.feature_importances_
# summarize feature importance
v=0
t=0
for i,v in enumerate(importance):
	 
    t = t + v
    
    print('Feature: %0d, Score: %.5f' % (i,v))
        
# plot de feature importance con la funcion Feature plot
feature_plot(importance,X_train,top_k=TOP_K)
```

* Las features que se aprecian en la grafica son las que representan cerca al 60% en importancia para poder predecir el target 

```python
# Con las más importantes features obtenidas en el paso anterior se crea un 
# datase más reducido y con ese se hace explora que clasificador es más optimo.

X_train_reduced = X_train[X_train.columns.values[(np.argsort(importance)[::-1])[:TOP_K]]]
X_test_reduced = X_test[X_test.columns.values[(np.argsort(importance)[::-1])[:TOP_K]]]
```

```python
# Con Random Forest calssifier
TOP_K =20
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
# define the model
RF_model = RandomForestClassifier(random_state = 0,n_estimators=50, max_depth= 10 )
# fit the model
RF_model.fit(X_train_reduced, y_train)
```

```python
# Con Decision Tree classifier 
TOP_K =20
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
# define the model
DT_model = DecisionTreeClassifier(random_state = 0, max_depth= 10 )
# fit the model
DT_model.fit(X_train_reduced, y_train)
```

```python
# Grado de exactitud a través del reporte de clasificación y análisis de la gráfica de Precision Recall.

from sklearn.metrics import accuracy_score,precision_recall_curve

y_pred_RF = RF_model.predict(X_test_reduced)
acc_RF = accuracy_score(y_test, y_pred_RF)

y_pred_DT = DT_model.predict(X_test_reduced) 
acc_DT = accuracy_score(y_test, y_pred_DT)
```

```python
#Precisión Random Forest
acc_RF
```

```python
#Precisión Decision tree
acc_DT
```

```python
# Precision Recall Curve para Random Forest
precision = dict()
recall = dict()
n_classes = len(set(df.SEMAFORO))
for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test[y_test.columns[i]],y_pred_RF[:,i])
    plt.plot(recall[i], precision[i], lw=2, label='class {}'.format(y.columns[i]))
    
plt.xlabel("recall")
plt.ylabel("precision")
plt.legend(loc="best")
plt.title("Random Forest precision vs. recall curve")
plt.show()
```

```python
# # Precision Recall Curve para Decision tree
precision = dict()
recall = dict()
n_classes = len(set(df.SEMAFORO))
for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test[y_test.columns[i]],y_pred_DT[:,i])
    plt.plot(recall[i], precision[i], lw=2, label='class {}'.format(y.columns[i]))
    
plt.xlabel("recall")
plt.ylabel("precision")
plt.legend(loc="best")
plt.title("Decision Tree precision vs. recall curve")
plt.show()
```

```python
# Matriz de Confusión para Random Forest
from sklearn.metrics import multilabel_confusion_matrix,classification_report,ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

cm = multilabel_confusion_matrix(y_test, y_pred_RF)

for i, confusion_matrix in enumerate(cm):
    disp = ConfusionMatrixDisplay(confusion_matrix)
    disp.plot(include_values=True, cmap="viridis", ax=None, xticks_rotation="vertical")
    plt.title('Class {}'.format(y.columns[i]))
    plt.show()


print(classification_report(y_test,y_pred_RF))  
```

```python
# Matriz de Confusión para Decision Tree
from sklearn.metrics import multilabel_confusion_matrix,classification_report,ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

cm = multilabel_confusion_matrix(y_test, y_pred_DT)

for i, confusion_matrix in enumerate(cm):
    disp = ConfusionMatrixDisplay(confusion_matrix)
    disp.plot(include_values=True, cmap="viridis", ax=None, xticks_rotation="vertical")
    plt.title('Class {}'.format(y.columns[i]))
    plt.show()


print(classification_report(y_test,y_pred_RF))  
```
